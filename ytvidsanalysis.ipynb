```python
#to see correlations between views and positive/negative engagement.

import matplotlib.pyplot as plt

# Views vs. Likes
plt.figure(figsize=(10, 6))
plt.scatter(vids['views'], vids['likes'], alpha=0.5)
plt.xlabel('Views')
plt.ylabel('Likes')
plt.title('Views vs. Likes')
plt.show()

# Views vs. Dislikes
plt.figure(figsize=(10, 6))
plt.scatter(vids['views'], vids['dislikes'], alpha=0.5)
plt.xlabel('Views')
plt.ylabel('Dislikes')
plt.title('Views vs. Dislikes')
plt.show()

# Correlation coefficients
print('Correlation between views and likes:', vids['views'].corr(vids['likes']))
print('Correlation between views and dislikes:', vids['views'].corr(vids['dislikes']))
```

```python
#to see correlation between views and discussion.

# Views vs. Comments
plt.figure(figsize=(10, 6))
plt.scatter(vids['views'], vids['comment_count'], alpha=0.5)
plt.xlabel('Views')
plt.ylabel('Comments')
plt.title('Views vs. Comments')
plt.show()

# Correlation coefficient
print('Correlation between views and comments:', vids['views'].corr(vids['comment_count']))
```

```python
#frequently used keywords to understand content themes
from wordcloud import WordCloud

# Combine title and tags into a single string
text = ""
for i in range(len(vids)):
  text += vids.loc[i]['title'] + " " + vids.loc[i]['tags']

# Generate word cloud
wordcloud = WordCloud(width=800, height=400).generate(text)

# Display the word cloud image
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

# Extract and print frequent words
words = wordcloud.words_
frequent_words = dict(sorted(words.items(), key=lambda item: item[1], reverse=True))

print("Frequent words:")
for word, frequency in frequent_words.items():
  print(f"{word}: {frequency}")
```
